```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head(10)
```

```python
movies = movies.merge(credits,how = "left",left_on='id',right_on = "movie_id")
```

```python
movies = movies.iloc[:2500,:]
```

```python
movies.columns
```

```python
movies = movies[["movie_id","title_x","overview","genres","keywords","cast","crew"]]
```

```python
movies.head(5)
```

```python
movies['genres'][0]
```

```python
def convert(obj):
  l = []
  for i in eval(obj):
    l.append(i['name'])
  return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(5)
```

```python
def convert2(obj):
  l = []
  a = 0
  for i in eval(obj):
    if a < 3:
      l.append(i['name'])
      a+=1
  return l
```

```python
movies['cast'] = movies['cast'].apply(convert2)
```

```python
movies['crew'][0]
```

```python
def fetch_director(a):
  l = []
  for i in eval(a):
    if i['job'] == 'Director':
      l.append(i['name'])
  return l
```

```python
movies['crew'] = movies['crew'].apply (fetch_director)
```

```python
movies.head(5)
```

```python
def gapremover(a):
  l = []
  for i in a:
    l.append(i.replace(" ",""))
  return l
```

```python
movies['cast'] = movies['cast'].apply(gapremover)
```

```python
movies['crew'] = movies['crew'].apply(gapremover)
```

```python
movies['genres'] = movies['genres'].apply(gapremover)
```

```python
movies.loc[:,'keywords'] = movies['keywords'].apply(gapremover)
```

```python
movies.head(5)
```

```python
movies.dropna(subset=['overview'],inplace=True)
```

```python
movies.loc[:,'overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies
```

```python
movie = movies.drop(columns = ['overview','genres','keywords','cast','crew'])
```

```python
movie
```

```python
movie.loc[:,'tags'] = movie['tags'].apply(lambda x:" ".join(x))
```

movie

```python
movie.head(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
movie.head(5)
```

```python
vector = cv.fit_transform(movie['tags']).toarray()
```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movies):
    index = movie[movie['title_x'] == movies].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(movie.iloc[i[0]].title_x)
```

```python
recommend('Avatar')
```
